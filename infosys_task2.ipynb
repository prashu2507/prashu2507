```python
import faiss
import numpy as np
from sentence_transformers import SentenceTransformer
```

```python
# Sample movie data (title and description)
movies = [
    {"title": "Inception", "description": "A thief who steals corporate secrets through the use of dream-sharing technology."},
    {"title": "The Matrix", "description": "A computer hacker learns about the true nature of his reality."},
    {"title": "Interstellar", "description": "A team of explorers travel through a wormhole in space."},
    {"title": "The Godfather", "description": "An organized crime dynasty's aging patriarch transfers control of his clandestine empire."},
    {"title": "The Dark Knight", "description": "When the menace known as the Joker emerges from his mysterious past."},
    {"title": "Pulp Fiction", "description": "The lives of two mob hitmen, a boxer, a gangster's wife, and a pair of diner bandits."},
    {"title": "Fight Club", "description": "An insomniac office worker and a devil-may-care soap maker form an underground fight club."}
]

# Prepare a list of movie descriptions for embedding
movie_descriptions = [movie["description"] for movie in movies]
```

```python
 # Load the pre-trained model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Convert movie descriptions to embeddings
movie_embeddings = model.encode(movie_descriptions).astype('float32')
```

```python
# Create a FAISS index
movie_index = faiss.IndexFlatL2(movie_embeddings.shape[1])

# Add movie embeddings to the index
movie_index.add(movie_embeddings)
```

```python
# Define a query movie description
query_description = "A skilled thief is given a chance at redemption if he can successfully perform inception."

# Convert the query description to an embedding
query_embedding = model.encode([query_description]).astype('float32')
```

```python
# Specify the number of similar results to retrieve
k = 3

# Search for the closest matches
distances, indices = movie_index.search(query_embedding, k)
```

```python
# Display the results
print("Query description:", query_description)
print("\nMost similar movies:")
for i, idx in enumerate(indices[0]):
    print(f"{i+1}: {movies[idx]['title']} - {movies[idx]['description']} (Distance: {distances[0][i]:.4f})")
```
